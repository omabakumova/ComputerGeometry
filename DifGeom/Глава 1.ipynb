```python


def plot_vector_function(r_function, t_range, title="Годограф векторной функции"):
  """
  Строит годограф векторной функции r(t).

  Args:
    r_function: Функция, которая принимает параметр t и возвращает вектор r(t).
    t_range: Диапазон значений параметра t.
    title: Заголовок графика.
  """

  # Генерируем значения t в заданном диапазоне
  t = np.linspace(t_range[0], t_range[1], 100)

  # Вычисляем координаты точек годографа
  x = [r_function(t_i)[0] for t_i in t]
  y = [r_function(t_i)[1] for t_i in t]
  z = [r_function(t_i)[2] for t_i in t]

  # Создаем 3D график
  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')
  ax.plot(x, y, z, label='Годограф')

  # Настройка графика
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_zlabel('Z')
  ax.set_title(title)
  ax.legend()
  plt.show()

# Функция r(t) = (cos(t), sin(t), t)
def r(t):
  return np.array([np.cos(t), np.sin(t), t])

# Диапазон значений параметра t
t_range = (0, 6 * np.pi)  # Измененный диапазон для лучшей визуализации

# Построение годографа
plot_vector_function(r, t_range, "Годограф функции r(t) = (cos(t), sin(t), t)")
```

```python


def plot_vector_function(r_function, t_range, title="Годограф векторной функции"):
    """
    Строит годограф векторной функции r(t).

    
        r_function: Функция, которая принимает параметр t и возвращает вектор r(t).
        t_range: Диапазон значений параметра t.
        Стрелочки у годографа используются для того, чтобы показать направление движения точки, которая описывает годограф, когда параметр `t` увеличивается.

    Вот как это работает:
    
    Вектор `r(t)`:  Векторная функция `r(t)` определяет положение точки в пространстве в зависимости от значения параметра `t`.
    Изменение `t`:  Когда `t` увеличивается, вектор `r(t)` также меняется, и точка перемещается по годографу.
    Стрелка:  Стрелка на годографе показывает направление движения точки, когда `t` увеличивается. 

    """

    # Генерируем значения t в заданном диапазоне
    t = np.linspace(t_range[0], t_range[1], 100)

    # Вычисляем координаты точек годографа
    x = [r_function(t_i)[0] for t_i in t]
    y = [r_function(t_i)[1] for t_i in t]
    z = [r_function(t_i)[2] for t_i in t]

    # Вычисляем направления векторов
    dx = [-np.sin(t_i) for t_i in t]
    dy = [np.cos(t_i) for t_i in t]
    dz = [1 for _ in t]

    # Создаем 3D график
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z, label='Годограф')

    # Добавляем стрелки, показывающие направление движения
    ax.quiver(x, y, z, dx, dy, dz, length=0.5, normalize=True, color='r')

    # Настройка графика
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(title)
    ax.legend()
    plt.show()

# Функция r(t) = (cos(t), sin(t), t)
def r(t):
    return np.array([np.cos(t), np.sin(t), t])

# Диапазон значений параметра t
t_range = (0, 6 * np.pi)  # Измененный диапазон для лучшей визуализации

# Построение годографа
plot_vector_function(r, t_range, "Годограф функции r(t) = (cos(t), sin(t), t)")
```

### Пример гладкой кривой(стр.17)


$$
y = \begin{cases} 
x^2 \cos\left(\frac{1}{x}\right), & \text{если } x \neq 0 \\
0, & \text{если } x = 0 
\end{cases}
$$


```python

"""

1. Для x ≠ 0, кривая будет представлять собой функцию вида y = x^2 * cos(1/x). 
Эта часть кривой будет гладкой, так как функция является непрерывной и дифференцируемой для всех x ≠ 0.

2. В точке x = 0, кривая будет иметь разрыв, так как функция x^2 * cos(1/x) не определена в этой точке. 
Вместо этого, в системе указано, что в точке x = 0 значение y = 0.

"""

def smooth_curve(x):
    if x != 0:
        return x**2 * np.cos(1/x)
    else:
        return 0

# Создание массива x-координат
x = np.linspace(-1, 1, 1000)

# Вычисление y-координат
y = [smooth_curve(xi) for xi in x]

# Построение графика
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('Гладкая кривая')
plt.show()
```

```python


# Определение параметрической кривой r(t)
def r(t):
    return np.array([np.cos(t), np.sin(t)])  # Пример: окружность

# Параметры
alpha = 0
beta = 2 * np.pi
n = 10  # Количество сегментов

# Разбиение сегмента [alpha, beta] на n частей
t_values = np.linspace(alpha, beta, n+1)

# Вычисление точек Mi на кривой L
points = np.array([r(t) for t in t_values])

# Построение ломаной, вписанной в кривую L
plt.plot(points[:, 0], points[:, 1], 'olivedrab', label='Ломаная M0M1...Mn')
plt.plot(points[:, 0], points[:, 1], 'ro')  # Отметка точек Mi

# Построение самой кривой L для сравнения
t_fine = np.linspace(alpha, beta, 1000)
curve = np.array([r(t) for t in t_fine])
plt.plot(curve[:, 0], curve[:, 1], 'tomato', label='Кривая L')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Ломаная, вписанная в кривую L')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

### Задача(стр.25).
Доказать, что для спрямляемости простой кривой \( L \) необходимо и достаточно, чтобы функции $\phi(t)$, $\psi(t)$, $\chi(t)$ были функциями ограниченной вариации.



Для доказательства того, что для спрямляемости простой кривой \( L \) необходимо и достаточно, чтобы функции $\phi(t)$, $\psi(t)$, $\chi(t)$  были функциями ограниченной вариации, можно воспользоваться следующими рассуждениями:

1. **Необходимость**:
   Пусть кривая \( L \) спрямляема. Это означает, что длина кривой \( L \) конечна. Длина кривой \( L \), заданной параметрически функциями $\phi(t)$, $\psi(t)$, $\chi(t)$  на интервале [a, b], вычисляется по формуле:
   $$
   L = \int_a^b \sqrt{\left(\frac{d\phi}{dt}\right)^2 + \left(\frac{d\psi}{dt}\right)^2 + \left(\frac{d\chi}{dt}\right)^2} \, dt
   $$
   Если длина кривой \( L \) конечна, то интеграл справа также конечен. Это означает, что функции $\phi(t)$, $\psi(t)$, $\chi(t)$ имеют конечные производные почти всюду на \([a, b]\), и их производные интегрируемы. Следовательно, функции $\phi(t)$, $\psi(t)$, $\chi(t)$ являются функциями ограниченной вариации.

2. **Достаточность**:
   Пусть функции $\phi(t)$, $\psi(t)$, $\chi(t)$ являются функциями ограниченной вариации на интервале [a, b]. Это означает, что их производные существуют почти всюду и интегрируемы. Тогда интеграл
   $$
   \int_a^b \sqrt{\left(\frac{d\phi}{dt}\right)^2 + \left(\frac{d\psi}{dt}\right)^2 + \left(\frac{d\chi}{dt}\right)^2} \, dt
   $$
   конечен, что означает, что длина кривой \( L \) конечна. Следовательно, кривая \( L \) спрямляема.

Таким образом, мы доказали, что для спрямляемости простой кривой \( L \) необходимо и достаточно, чтобы функции $\phi(t)$, $\psi(t)$, $\chi(t)$ были функциями ограниченной вариации.

Соприкасающаяся плоскость в точке $M$ кривой $l$ — это плоскость, имеющая с $l$ в точке $M$ касание порядка $n \geq 2$ (см. Соприкосновение). Соприкасающаяся плоскость может быть также определена как предел переменной плоскости, проходящей через три точки кривой $l$, когда эти точки стремятся к точке $M$. С механической точки зрения соприкасающаяся плоскость может быть охарактеризована как плоскость ускорений: при произвольном движении материальной точки по кривой $l$ вектор ускорения лежит в соприкасающейся плоскости. Обычно кривая, кроме исключительных случаев, пронизывает свою соприкасающуюся плоскость в точке соприкосновения (см. рис.).

Если кривая $l$ задана уравнениями $x = x(u)$, $y = y(u)$, $z = z(u)$, то уравнение соприкасающейся плоскости имеет вид:
$$
\begin{vmatrix}
X - x & Y - y & Z - z \\
x' & y' & z' \\
x'' & y'' & z''
\end{vmatrix} = 0
$$
где $X, Y, Z$ — текущие координаты, а $x, y, z, x', y', z', x'', y'', z''$ вычисляются в точке соприкосновения. Если все три коэффициента при $X, Y, Z$ в уравнении соприкасающейся плоскости исчезают, то соприкасающаяся плоскость становится неопределённой (может совпадать с любой плоскостью, проходящей через касательную).




```python


# Задаем уравнения кривой
def curve(u):
    x = u
    y = u**2
    z = u**3
    return x, y, z

# Создаем функцию для построения графика
def plot_tangent(elev, azim):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Вычисляем уравнение соприкасающейся плоскости
    u_m = 2
    x_m, y_m, z_m = curve(u_m)
    
    x_prime = 1
    y_prime = 2*u_m
    z_prime = 3*u_m**2
    x_double_prime = 0
    y_double_prime = 2
    z_double_prime = 6*u_m
    
    A = y_prime*z_double_prime - z_prime*y_double_prime
    B = z_prime*x_double_prime - x_prime*z_double_prime
    C = x_prime*y_double_prime - y_prime*x_double_prime
    D = -(A*x_m + B*y_m + C*z_m)
    
    u = np.linspace(-5, 5, 50)
    v = np.linspace(-5, 5, 50)
    U, V = np.meshgrid(u, v)
    X = U
    Y = V
    Z = -(A*U + B*V + D) / C
    
    ax.plot(curve(np.linspace(-5, 5, 100))[0], curve(np.linspace(-5, 5, 100))[1], curve(np.linspace(-5, 5, 100))[2], color='b')
    surface = ax.plot_surface(X, Y, Z, alpha=0.5)
    point = ax.scatter(x_m, y_m, z_m, color='r')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    ax.view_init(elev=elev, azim=azim)
    
    plt.show()

# Создаем виджеты для управления углом обзора
elev_slider = widgets.FloatSlider(value=30, min=0, max=90, step=1, description='Elevation:')
azim_slider = widgets.FloatSlider(value=0, min=0, max=360, step=1, description='Azimuth:')
widgets.interactive(plot_tangent, elev=elev_slider, azim=azim_slider)
```

Для составления уравнений касательной, главной нормали и бинормали кривой, а также нормальной и спрямляющей плоскостей, нам нужно знать параметрическое уравнение кривой. Пусть кривая задана параметрически как $\mathbf{r}(t) = (x(t), y(t), z(t))$.

1. Касательная вектор:
   $$
   \mathbf{T}(t) = \frac{\mathbf{r}'(t)}{\|\mathbf{r}'(t)\|}
   $$
   где $\mathbf{r}'(t) = \left( \frac{dx}{dt}, \frac{dy}{dt}, \frac{dz}{dt} \right)$.

2. Главная нормаль:
   $$
   \mathbf{N}(t) = \frac{\mathbf{T}'(t)}{\|\mathbf{T}'(t)\|}
   $$
   где $\mathbf{T}'(t) = \left( \frac{dTx}{dt}, \frac{dTy}{dt}, \frac{dTz}{dt} \right)$.

3. Бинормаль:
   $$
   \mathbf{B}(t) = \mathbf{T}(t) \times \mathbf{N}(t)
   $$
   где $\times$ обозначает векторное произведение.

4. Нормальная плоскость:
   Уравнение нормальной плоскости в точке $\mathbf{r}(t0)$ имеет вид:
   $$
   (\mathbf{r} - \mathbf{r}(t0)) \cdot \mathbf{T}(t0) = 0
   $$
   где $\mathbf{r} = (x, y, z)$.

5. Спрямляющая плоскость:
   Уравнение спрямляющей плоскости в точке $\mathbf{r}(t0)$ имеет вид:
   $$
   (\mathbf{r} - \mathbf{r}(t0)) \cdot \mathbf{B}(t0) = 0
   $$

Пример:
Пусть кривая задана как $\mathbf{r}(t) = (t, t^2, t^3)$.

1. Касательная вектор:
   $$
   \mathbf{r}'(t) = (1, 2t, 3t^2)
   $$
   $$
   \mathbf{T}(t) = \frac{(1, 2t, 3t^2)}{\sqrt{1 + 4t^2 + 9t^4}}
   $$

2. Главная нормаль:
   $$
   \mathbf{T}'(t) = \left( 0, \frac{2(1 + 9t^4)}{(1 + 4t^2 + 9t^4)^{3/2}}, \frac{6t(1 + 4t^2)}{(1 + 4t^2 + 9t^4)^{3/2}} \right)
   $$
   $$
   \mathbf{N}(t) = \frac{\mathbf{T}'(t)}{\|\mathbf{T}'(t)\|}
   $$

3. Бинормаль:
   $$
   \mathbf{B}(t) = \mathbf{T}(t) \times \mathbf{N}(t)
   $$

4. Нормальная плоскость:
   $$
   (x - t) + 2t(y - t^2) + 3t^2(z - t^3) = 0
   $$

5. Спрямляющая плоскость:
   $$
   (x - t)Bx + (y - t^2)By + (z - t^3)Bz = 0
   $$

Эти уравнения позволяют описать касательную, главную нормаль и бинормаль кривой, а также нормальную и спрямляющую плоскости.

### Прежде чем пояснить, почему порядок равен 37, укажу основную формулу, по которой находится сам порядок, и необходимые условия.(стр.42)

Порядок соприкосновения кривых $L1$ и $L2$ в точке $M0$ не ниже $q$ $∈ N$, если:

$$\lim{l \to 0} \frac{h}{l^q} = 0$$

Если, кроме того,

$$\lim{l \to 0} \frac{h}{l^{q+1}} \neq 0$$

или отношение $\frac{h}{l^{q+1}}$ не имеет предела при $l \to 0$, то будем говорить, что порядок соприкосновения кривых $L1$ и $L2$ в точке $M0$ равен $q$.



Для определения порядка соприкосновения кривых $L1$ и $L2$ в точке $(0, 0)$, мы должны рассмотреть функции $h(t)$ и $l(t)$ и их отношение.

Функция $h(t)$ у нас:
$$ h(t) = |t|^{37.2} \sqrt{1 + |t|^{105.8}} $$

Функция $l(t)$ у нас:
$$ l(t) = |t| $$

Теперь нам нужно вычислить предел:
$$ \lim{t \to 0^+} \frac{h(t)}{l(t)^{37}} $$

Подставим функции $h(t)$ и $l(t)$ в этот предел:
$$ \lim{t \to 0^+} \frac{|t|^{37.2} \sqrt{1 + |t|^{105.8}}}{|t|^{37}} $$

Так как $t$ положительно, то $|t| = t$. Следовательно, наш предел становится:
$$ \lim{t \to 0^+} \frac{t^{37.2} \sqrt{1 + t^{105.8}}}{t^{37}} $$

Упростим выражение:
$$ \lim{t \to 0^+} t^{0.2} \sqrt{1 + t^{105.8}} $$

Теперь рассмотрим поведение функции $t^{0.2} \sqrt{1 + t^{105.8}}$ при $t \to 0^+$. Поскольку $t^{105.8}$ стремится к нулю гораздо быстрее, чем $t^{0.2}$, мы можем приблизительно считать $\sqrt{1 + t^{105.8}} \approx 1$ при $t \to 0^+$.

Таким образом, предел становится:
$$ \lim{t \to 0^+} t^{0.2} \cdot 1 = \lim{t \to 0^+} t^{0.2} = 0 $$

Это подтверждает, что порядок соприкосновения кривых $L1$ и $L2$ в точке $(0, 0)$ равен 37, так как $t^{0.2}$ стремится к нулю при $t \to 0^+$.

Теперь рассмотрим предел отношения $\frac{h(t)}{l(t)^{37+1}}$:
$$ \lim{t \to 0^+} \frac{h(t)}{l(t)^{38}} = \lim{t \to 0^+} \frac{|t|^{37.2} \sqrt{1 + |t|^{105.8}}}{|t|^{38}} = \lim{t \to 0^+} \frac{t^{37.2} \sqrt{1 + t^{105.8}}}{t^{38}} $$

Упростим выражение:
$$ \lim{t \to 0^+} t^{-0.8} \sqrt{1 + t^{105.8}} $$

Поскольку $t^{-0.8}$ стремится к бесконечности при $t \to 0^+$, предел не существует (он равен бесконечности).

Таким образом, порядок соприкосновения кривых $L1$ и $L2$ в точке $(0, 0
)$ равен 37, так как предел отношения $\frac{h(t)}{l(t)^{37+1}}$ не существует (он равен бесконечности).
