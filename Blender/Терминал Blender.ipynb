## Использование терминала Python в Blender для взаимодействия с объектами

#### В дополнение к текстовому редактору, который позволяет писать скрипты и взаимодействовать со средой Blender, существует также терминал (командная строка) для непосредственного взаимодействия с объектами и их создания.

#### В данном случае мы рассмотрим базовые команды, которые можно использовать в терминале для взаимодействия с объектами в Blender.

````python
from IPython.display import Image, display
````

### 1) Изменение цвета.
#### С помощью терминала мы можем изменить цвет созданного нами объекта.Используя следующую команду:
---
> ```python
> import bpy
> bpy.data.objects["Pyramid"].active_material.diffuse_color = (1.0, 0.0, 0.0, 1.0)
---
#### Пирамида станет красного цвета.

````python
Image('3.jpg') 
````

### 2) Изменение положения объекта.
#### Также возможно менять положение объекта на определенное положение:
---
> ```python
> bpy.data.objects["Cube"].location = (1.0, 2.0, 3.0)
---


````python
Image('4.jpg') 
````

## 3) Добавление новых объектов в среду.
#### Помимо существующего мы также можем с помощью терминала создать новый:
---
> ```python
> bpy.ops.mesh.primitive_cube_add()
---
#### В качестве примера,я создала куб.

````python
Image('5.jpg') 
````

#### Теперь мы можем также взаимодействовать с новым объектом через терминал.

## 4) Добавление модификаторов объекта.
#### Мы можем воздействовать на объект с помощью модификаторов,меняя его внешний вид более явно.
---
> ```python
> bpy.data.objects["Куб"].modifiers.new("Subsurf", type='SUBSURF')
---
#### В данном случае,я применила модификатор к объекту "Куб".

````python
Image('6.jpg') 
````

#### Модификаторы бывают разные,мы можем использовать иной вариант того же модификатора.
---
> ```python
> bpy.data.objects["Куб"].modifiers["Subsurf"].levels = 2
---


````python
Image('7.jpg') 
````

## 5) Добавление света.
#### Мы также помимо объектов можем добавить свет(точечный источник).
---
> ```python
> bpy.ops.object.light_add(type='POINT')
---
#### Он также появится в нашей коллекции рядом с другими объектами.

````python
Image('8.jpg') 
````

#### Можно получить координаты вершин объекта,используя следующую команду:
---
> ```python
> bpy.data.objects["Куб"].data.vertices[0].co
---
#### Мы получим следующий вывод в терминале:

````python
Image('9.jpg') 
````

#### Можем также получить количество вершин объекта:
---
> ```python
> len(bpy.data.objects["Куб"].data.vertices)

## 9) Анимация.
#### Самое интересное,что можно выдлеить из базовых команд - это возможность анимировать объект.
---
> ```python
> bpy.context.object.location = (0.0, 0.0, 0.0)
> bpy.context.object.keyframe_insert(data_path="location", index=-1)
> bpy.context.scene.frame_set(10) # Переход к кадру 10
> bpy.context.object.location = (2.0, 0.0, 0.0)
> bpy.context.object.keyframe_insert(data_path="location", index=-1)
---
#### Гифка конечно не очень,но здесь продемонстрирована самая простая анимация - движение по оси объекта.

#### Также анимацию можно включать:
---
> ```python
> bpy.context.object.animation_data.action.fcurves[0].mute = False

#### И выключать:
---
> ```python
> bpy.context.object.animation_data.action.fcurves[0].mute = True
